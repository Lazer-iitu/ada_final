```python
rename_dict = {
    "Timestamp": "response_timestamp",
    "Ваш возраст": "age",
    "Ваш пол": "gender",
    "Ваш ежемесячный доход": "monthly_income_range",
    "Ваш доход стабильный?": "income_stability",
    "Примерно сколько вы тратите в месяц? (в тенге)": "total_monthly_spending",
    "Сколько вы тратите на еду в месяц (в тенге)?": "food_spending",
    "Сколько вы тратите на транспорт в месяц?": "transport_spending",
    "Сколько вы тратите на развлечения в месяц?": "entertainment_spending",
    "Сколько вы тратите на покупки в месяц?": "shopping_spending",
    "Сколько вы тратите на аренду жилья в месяц?": "rent_spending",
    "Сколько вы тратите на другие ежемесячные расходы": "other_spending",
    "Отслеживаете ли вы свои расходы?": "expense_tracking",
    "Следуете ли вы ежемесячному бюджету? ": "budget_adherence",
    "Как часто вы совершаете импульсивные покупки?": "impulse_purchase_frequency",
    "Импульсивные покупки чаще происходят": "impulse_purchase_trigger",
    " Откладываете ли вы деньги каждый месяц?": "saves_monthly",
    "Если да, сколько вы накопили в прошлом месяце? (в тенге)": "amount_saved_last_month",
    "Какова ваша цель по накоплениям на месяц?": "monthly_savings_goal",
    "Насколько сложно вам откладывать деньги?": "saving_difficulty_score",
    "Какими финансовыми приложениями вы пользуетесь регулярно? ": "financial_apps_used",
    "Что является для вас самой большой сложностью при попытке откладывать деньги?": "main_saving_obstacle",
    "Если вы НЕ откладываете деньги, то почему?": "reason_for_not_saving"
}

df = df.rename(columns=rename_dict)
```

## **Initial data review**

```python
df.head()
```

```python
df = df.drop(columns=["response_timestamp", "financial_apps_used"])
```

```python
df.info()
```

```python
numeric_cols = [
    "total_monthly_spending", "food_spending", "transport_spending",
    "entertainment_spending", "shopping_spending", "rent_spending",
    "other_spending"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
categorical_cols = [
    "gender", "age", "monthly_income_range",
    "income_stability", "expense_tracking",
    "budget_adherence",
    "impulse_purchase_trigger",
    "reason_for_not_saving"
]
df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
df.to_csv("/content/drive/MyDrive/Colab Notebooks/ada/financial_habit.csv", index=False)
```
